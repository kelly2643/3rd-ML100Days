```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
matrix = np.random.uniform(-1,1,[10,10])
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(matrix,cmap = plt.cm.RdYlBu_r, vmin = -0.25, annot = True, vmax = 0.6)
plt.title('Correlation Heatmap')
```

```python
matrix1 = np.random.uniform(-1,1,[1000,3])
```

```python
# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=1000)
plot_data = pd.DataFrame(matrix1, indice).reset_index()
```

```python
plot_data.head()
```

```python
# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False,
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])
# 上半部為 scatter
grid.map_upper(plt.scatter, alpha = 0.2)

# 對角線畫 histogram
grid.map_diag(sns.kdeplot)

# 下半部放 density plot
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)
```
