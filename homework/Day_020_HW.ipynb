```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
matrix = np.random.uniform(-1, 1, (10, 10))
plt.figure(figsize=(10,10))
# 繪製相關係數 (correlations) 的 Heatmap

sns.heatmap(matrix, cmap = plt.cm.RdYlBu_r, annot = True)
plt.title('Correlation Heatmap'); 
plt.show()
```

```python
matrix =np.random.uniform(-1, 1, (1000, 3)) 
```

```python
plot_data = pd.DataFrame(matrix,index=list(range(1000)))
```

```python
plot_data['index']=list(range(1000))
```

```python
# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False, 
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])

# 上半部為 scatter
grid.map_upper(plt.scatter, alpha = 0.2)

# 對角線畫 histogram
grid.map_diag(sns.kdeplot)

# 下半部放 density plot
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.suptitle('Pairs Plot', size = 32, y = 1.05)
plt.show()
```
